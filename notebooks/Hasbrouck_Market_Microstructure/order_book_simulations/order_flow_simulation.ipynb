```python
import numpy as np
import matplotlib.pyplot as plt

from trader import Trader
from order_book import OrderBook

import utilities
```

```python
# initialise the trader and the order book
# here we will make just one trader make all the transactions
trader = Trader()

book = OrderBook()

np.random.seed(70)
```

```python
tick_size = 1
simulation_length = 50


# new we want to generate some order book snapshot.
# to do this, let's first fill the order book with some random trades
book.asks = [(101, 5), (102, 7), (103,2)]
book.bids = [(99, 5), (98, 7), (97,2)]

mid_price = book.return_mid_price()

book.print_order_book_state()

#book_state_sequence.append(append_book_state(book, 0))
#mid_price_sequence.append(mid_price)
#microprice_sequence.append(book.return_micro_price())


# now let's do some random trades
# in this example, we set the quantity with a uniform distribution from 1 to 5.

while book.time < simulation_length:
    # step 1: the market maker checks if the current spread is equal to its target spread
    # if it is -> it does nothing
    # if it is not -> add limit orders
    # TODO: track how many shares it is trading and try to keep a fixed inventory
    # TODO: the MM tries to go where the potential liquidity is, i.e. sets its limit prices
    # following the volumes in the book
    # se il mercato sembra in crescita -> piazzare limit buy
    # se il mercato sembra in calo -> limit sell
    # può usare ad esempio l'imbalance. segnale di salita se volume bid > volume ask
    # ecc...
    # alto volume su primo bid -> molti ordine in acquisto -> limit buy a bid più alto viene eseguito rapidamente
    # analogamente con ask, ma più basso
    
    # update the mid price
    mid_price = book.return_mid_price()

    order_type = np.random.choice(['limit_sell', 'limit_buy', 'market_buy', 'market_sell'])
    if order_type  == 'limit_sell':
        price = int(mid_price + np.random.normal(0,1) + tick_size) // tick_size
    elif order_type == 'limit_buy':
        price = int(mid_price - np.random.normal(0,1) - tick_size) // tick_size
    else:
        price = None

    quantity = int(np.random.uniform(low=1, high=5))
                   
    trader.submit_order_to_order_book(
        order_type=order_type, 
        price=price, 
        quantity=quantity,
        book=book)


    book.print_order_book_state()

```
